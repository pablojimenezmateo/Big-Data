# Comparing speedups when using compression with blz btables stored in disk

In this tutorial we are going to see the performance of the btable container using different compressors and compression levels. Note that all the containers are going to be in disk and the caches will be dropped every iteration, so everything is read from disk. We are going to use the same dataset that we used in "Querying big datasets, the good way" notebook.

**WARNING:** This notebook has been done under administrator privileges in order to erase the cache, so all the reading is done from the disk. **Do not** run this as administrator if you are not completely sure what is that and the risks you are exposed to.
```python
import pandas as pd
import numpy as np
import pylab as plt
import blz
import sys
import csv
from time import time
from shutil import rmtree
from collections import defaultdict
from subprocess import call
```

First, we read the dataset and store it as an uncompressed btable.

```python
t1 = time()
df = pd.read_hdf('h5/msft-18-SEP-2013-blosc9.h5', '/msft')
t2 = time()

print "Reading h5: " + str(t2 - t1)

dt = np.dtype([(k, v if v.kind!='O' else 'S49') for k,v in df.dtypes.iteritems()])

#If the blz already exist, remove it
rmtree('blz/msft-18-SEP-2013-blosc9.blz', ignore_errors=True)

t1 = time()
bt = blz.fromiter((i[1:] for i in df.itertuples()), rootdir='blz/msft-18-SEP-2013-blosc9.blz', dtype=dt, count=len(df), 
                  bparams=blz.bparams(clevel=0))
t2 = time()

print "Converting h5 to btable: " + str(t2 - t1)
```

Let's create a function that will copy the blz.

```python
def copy(src, dest, clevel=5, shuffle=True, cname="blosclz"):

    """

    Parameters
    ----------
    clevel : int (0 <= clevel < 10)
        The compression level.
    shuffle : bool
        Whether the shuffle filter is active or not.
    cname : string ('blosclz', 'lz4', 'lz4hc', 'snappy', 'zlib', others?)
        Select the compressor to use inside Blosc.

    """

    src = blz.btable(rootdir=src)
    
    copied = src.copy(rootdir=dest, bparams=blz.bparams(clevel=clevel, shuffle=shuffle, cname=cname))
    copied.flush()
```

We will take the [Open-High-Low-Close](http://en.wikipedia.org/wiki/Open-high-low-close_chart) points for that chart, for every level, for every compressor.

```python
def get_OHLC_points(src):
    
    #This two lines is why we need administrator privileges
    #This will only work under Linux
    call(["sync"])
    call(["echo", "3", ">", "/proc/sys/vm/drop_caches"])
    
    windows = []
    window_size = 79
    sample_size = 2455
    
    iter = src.where('(Type == \'Trade\')', 'Price')
    
    for i in xrange(sample_size):
    
        local_max = - np.inf
        local_min = np.inf
        average_price = 0
    
        for j in xrange(window_size):
        
            current = iter.next()
            price = current[0]
            average_price += price
            
            if(price > local_max):
                local_max = price
            
            if(price < local_min):
                local_min = price
            
            if(j == 0):
                local_open = price
            
            if(j == (window_size - 1)):
                local_close = price
        
        windows.append([average_price/float(window_size), local_open, local_max, local_min, local_close]) 
```

And now we just need a benchmark function.

```python
def benchmark(src, cmethods):

    for method in cmethods:

        myfile = open('csv/btable_' + method + '.csv', 'wb')
        wr = csv.writer(myfile, quoting=csv.QUOTE_ALL)
        wr.writerow(['Compression level',
                     'Compressed size',
                     'Compression ratio',
                     'Compression time',
                     'OHLC time',
                     'Writing speed (MiB/s)',
                     'OHLC speed (MiB/s)'])
        
        for compression_level in xrange(0,10):

            #I get the original file and compress it
            tc1 = time()
            copy(src, 'blz/temp.blz', compression_level, True, method)
            tc2 = time()

            #Get disk size
            data = blz.btable(rootdir='blz/temp.blz')
            
            #Uncompressed size in MiB
            uncompressed = data.nbytes/(2**20)
            
            #Now I get the poins and measure the time
            t1 = time()
            get_OHLC_points(data)
            t2 = time()

            #I should store stuff
            row = [compression_level, data.cbytes,
                   round(data.nbytes/float(data.cbytes),3),
                   str(tc2 - tc1),
                   str(t2 - t1),
                   uncompressed/(tc2 - tc1),
                   uncompressed/(t2 - t1)]

            #Add it to the csv
            wr.writerow(row)
            rmtree('blz/temp.blz', ignore_errors=True)

        myfile.close()
```

We have all the ingredients now, let's do some benchmarking.

```python
src = 'blz/msft-18-SEP-2013-blosc9.blz'
cmethods = ['blosclz', 'lz4', 'lz4hc', 'snappy', 'zlib']

t1 = time()
benchmark(src, cmethods)
t2 = time()

print t2-t1
```

Let's do some more plotting. For that I first need to make some dictionaries out of this data.

```python
def get_dict(filename):
    
    columns = defaultdict(list)
    with open('csv/btable_' + filename) as f:
        reader = csv.reader(f)
        reader.next()
        for row in reader:
            for (i,v) in enumerate(row):
                if i == 1:
                    columns[i].append(float(v)/131072)
                    continue
                columns[i].append(v)
    return columns

blosclz = get_dict('blosclz.csv')
lz4 = get_dict('lz4.csv')
lz4hc = get_dict('lz4hc.csv')
snappy = get_dict('snappy.csv')
zlib = get_dict('zlib.csv')
```

We here can see that except with zlib, it is actually better to compress the data rather than to have it uncompressed (Better computational times)
